```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
boston_dataset = datasets.load_boston()
X_full = boston_dataset.data
Y = boston_dataset.target
print X_full.shape
print Y.shape
```

```python
selector = SelectKBest(f_regression, k=1)
selector.fit(X_full, Y)
X = X_full[:,selector.get_support()]
print X.shape
```

```python
plt.scatter(X, Y, color='blue')
plt.show()
```

```python
regressor = LinearRegression(normalize=True)
regressor.fit(X, Y)
plt.scatter(X, Y, color='green')
plt.plot(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```

```python
regressor = SVR()
regressor.fit(X, Y)
plt.scatter(X, Y, color='black')
plt.scatter(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```

```python
regressor = RandomForestRegressor()
regressor.fit(X, Y)
plt.scatter(X, Y, color='black')
plt.scatter(X, regressor.predict(X), color='blue', linewidth=3)
plt.show()
```

Use pandas and numpy to build a scatter matrix

```python
import pandas as pd
import numpy as np
```

```python
colors = list()
palette = {0: "red", 1: "green", 2: "blue"}
for c in np.nditer(iris.target): colors.append(palette[int(c)])
```

Using the paletter dictionary convert each numeric to color string

```python
dataframe = pd.DataFrame(iris.data,
                         columns=iris.feature_names)
```

```python
scatterplot = pd.scatter_matrix(dataframe,
                                alpha=0.3,
                                figsize=(10, 10),
                                diagonal='hist',
                                c=colors,
                                marker='o',
                                grid=True)
```
